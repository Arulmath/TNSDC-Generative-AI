```python
import cv2

# Load the two images
image1 = cv2.imread('/content/nature.jpg')
image2 = cv2.imread('/content/Lipbalm.jpg')

# Resize the second image to match the size of the first image
image3 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Blending the two images
alpha = 0.5
blended_image = cv2.addWeighted(image1, alpha, image3, 1 - alpha, 0)

# Display the blended image
cv2_imshow(blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np


# Read the 4 images
img1 = cv2.imread('/content/nature.jpg')
img2 = cv2.imread('/content/Lion.jpg')
img3 = cv2.imread('/content/cow.jpg')
img4 = cv2.imread('/content/Tiger.jpg')

# Resize all images to the same dimensions
width = 500
height = 500
img5 = cv2.resize(img1, (width, height))
img6 = cv2.resize(img2, (width, height))
img7 = cv2.resize(img3, (width, height))
img8 = cv2.resize(img4, (width, height))

# Create weight matrices for each image
alpha = 0.25
beta = 0.5
gamma = 0.5
delta = 0.25

# Blend the images
blended_img1 = cv2.addWeighted(img5, alpha, img6, beta, 0)
blended_img2 = cv2.addWeighted(blended_img1, 1, img7, gamma, 0)
Final_img = cv2.addWeighted(blended_img2, 1, img8, delta, 0)

# Display the final blended image
cv2_imshow(Final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
